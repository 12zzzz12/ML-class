```python
# 라이브러리 import

import numpy as np                
import matplotlib.pyplot as plt  
import pandas as pd               
import mglearn 
```

```python
#sklearn.dataset에서 load_iris 데이터를 import 한다.
from sklearn.datasets import load_iris
```

```python
#iris_dataset 이라는 변수에 load_iris 저장

iris_dataset = load_iris()
```

```python
#iris_dataset의 key를 확인한다.

print("iris_dataset의 키: {}".format(iris_dataset.keys()))
```

```python
#DESCR을 통해 iris_dataset의 정보를 간략하게 설명해줌.
#인스턴스 개수와 어트리뷰트의 개수, 도메인 출력

print(iris_dataset["DESCR"][:193]+"\n....")
```

```python
#iris_dataset의 정보를 상세하게 프린트한다.
#attribute, class 등의 정보를 알 

print(iris_dataset.DESCR)
```

```python
#iris_dataset의 target_name들을 출력한다. 

print("타깃의 이름: {}".format(iris_dataset['target_names']))
```

```python
#iris_dataset의 feature_name들을 출력한다.

print("특징이름:{}".format(iris_dataset['feature_names']))
```

```python
#iris_dataset의 target들의 정보를 출력한다 3가지의 품종을 0,1,2를 통해 표현한다.

print("타겟:{}".format(iris_dataset['target']))
```

```python
#iris_dataset의 모든 튜플의 data를 출력한다. 4가지 속성을 가진 150개의 튜플 출력
#위의 순서대로 출력을 하므로 근접한 data들은 비슷한 값을 가지는 것을 확인할 수 있다.

print("데이터:{}".format(iris_dataset['data']))
```

```python
#dataset의 data를 처음 5행까지 출력한다.

print("처음 5행:\n{}".format(iris_dataset['data'][:5]))
```

```python
#iris_dataset의 data 타입을 출력한다. numpy.ndarray 타입임을 확인할 수 있다.

print("data의 타입:{}".format(type(iris_dataset['data'])))
```

```python
#iris_dataset의 data크기를 출력한다. 150개의 튜플이 4개의 속성을 가짐을 확인할 수 있다.

print("data의 크기:{}".format(iris_dataset['data'].shape))
```

```python
#sklearn.modelselection에서 train_test_split을 import한다. 
#train_split함수를 이용해서 X_train,  X_test, y_train, y_test 로 data를 훈련할
#부분과 테스트 할 부분으로 75:25 비율로 나눈다. 
#random_state는 dataset을 나눌 때 항상 같은 값으로 나누어지도록 난수를 설정하는 것이다.

from  sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
#X_train은 data 값 (5, 2, 4.0, 1) 이런 거 들어있다. 훈련 데이터 (행렬이라 대문자)
#y_train은 target값이 들어있다. (0, 1, 2)                       (백터라 소문자)
#X_train(data값) == y_train(target값)
print("X_train의 크기:{}".format(X_train.shape))
print("y_train의 크기:{}".format(y_train.shape))
```

```python
#X_train의 개수를 뺀 38개의 data 값이 들어있다. 
#y_train의 개수를 뺀 38개의 target 값이 들어있다.
#X_test(data값) == y_test(target값)

print("X_test의 크기:{}".format(X_test.shape))
print("y_test의 크기:{}".format(y_test.shape))
```

```python
#(X_train)  X_train data를 이용해서 데이터 프레임을 생성한다.
#(columns = iris da) 열의 이름은 iris_dataset.feature_names에 있는 문자열을 사용한다. 

iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
```

```python
#데이터 프레임을 사용해 y_train에 따라 색으로 구분된 산점도 행렬을 생성한다.

# figsize = 그림의 크기 (가로, 세로)
# maker = 점의 모양 (o = 원, s = 사각형)
# s = marker의 크기
# alpha = 투명성 정도

pd.plotting.scatter_matrix(iris_dataframe, c= y_train, figsize=(15,15), marker = 'o',
                          hist_kwds = {'bins':20}, s= 60, alpha=8, cmap=mglearn.cm3)
```

```python
#sklearn.neighbors 에서 knn 분류기를 import 한다.
#knn알고리즘을 이웃의 수를 1로 설정한다.
#위에서 나누어놓은 X_train과 y_train을 훈련시킨다.

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
#X_new 라는 새로운 배열을 생성하여 해당 데이터 값을 저장한다.
#X_new.shape를 통해 (1,4)의 배열이 생성된 것을 확인하였다.

X_new = np.array([[5, 2.9, 1 ,0.2]])
print("X_new.shape:",X_new.shape)
```

```python
#위에서 실행한 knn 학습을 통해 X_new를 집어넣고 예측값을 저장한다.
#X_new의 target 값을 출력한다. (0,1,2) 중에 하나 
#예측한 타깃의 이름을 target_names를 이용해 출력한다. (target =0)일때 class 이름

prediction = knn.predict(X_new)
print("예측:",prediction)
print("예측한 타깃의 이름:", iris_dataset['target_names'][prediction])
```

```python
#위에서 테스트하기 위해 남겨 놓은 X_test값을 knn에 집어넣어 데이터들의 target값을 얻는다.
#X_test의 데이터 target값이 (0,1,2)로 잘 나누어진것을 확인할 수 있다.

y_pred= knn.predict(X_test)
print("테스트 세트에 대한 예측값\n", y_pred)
```

```python
#위에서 y_test로 미리 빼놓은 target값들과 
#방금 knn(X_test)를 통해 얻은 y_pred의 target값을 
#비교하면 정확도가 어느정도 인지 확인할 수 있다.

print("테스트 세트의 정확도:{:.2f}".format(np.mean(y_pred == y_test)))
```

```python
#knn.score를 통해 X_test와 y_test 값을 집어넣어 정확도를 확인할 수 있다.

print("테스트 세트의 정확도:{:.2f}".format(knn.score(X_test , y_test)))
```
