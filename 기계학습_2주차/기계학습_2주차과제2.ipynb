```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import mglearn
```

```python
#한국어를 출력할 때 꺠짐 현상을 없애기 위한 구문이다.

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
#make_forge() 메소드를 이용해 X, y에 데이터 값들을 저장한다.
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
```

```python
#산점도를 그리기 위한 구문
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.legend(["클래스 0","클래스 1"], loc=4) #클래스를 설명하기 위한 그림과 위치
plt.xlabel("first feature") #첫번째 feature라는 의미를 x축에 출력
plt.ylabel("second feature") #두번째 feature라는 의미를 y축에 출력
print("X.shape:",X.shape) #X의 형태(칼럼,피처)를 출력
```

```python
#X의 형태(튜플,피처) , y의 형태(튜플,피처)를 출력
print("X.shape: {}".format(X.shape), "y.shape:{}".format(y.shape))
```

```python
#train_test_split을 이용해 훈련, 테스트 데이터를 나눈다.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
#X_train : X_test = 75:25 비율로 나뉘어진걸 확인할 수 있다.
print("X_train.shape:{}".format(X_train.shape), "X_test.shape:{}".format(X_test.shape))
```

```python
#최근접 알고리즘을 clf에 저장한다. 이웃수는 3으로 지정한다.
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
#clf를 이용해 X_train과 y_train을 훈련시킨다.
clf.fit(X_train, y_train)
```

```python
#X_test값을 예측한 값을 출력한다.
print("테스트 세트 예측:{}".format(clf.predict(X_test)))
```

```python
#X_test와 y_test값을 이용해 테스트의 정확도를 출력한다.
print("테스트 세트 정확도:{:.2f}".format(clf.score(X_test, y_test)))
```

```python
#fig= figure(전체 subplot 몇개의 그래프가 있든 상관없이 그걸 모두 담는 하나)
#axe = 전체 중 낱낱개 (subplot안에 (a1, a2,a3 )의 그래프가 있다면 a1,a2,a3를 말함)

#subplots(행, 열, figsize(x, y) 크기를 말함)
fig, axes = plt.subplots(1, 3, figsize=(10,3))

#neighbors가 1, 3, 9일 때 각각의 그래프를 그림 
for n_neighbors, ax in zip([1,3,9], axes):
    
    #clf에 최근접 알고리즘 적용 이웃수는 1,3,9일 때
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    
    #점을 찍는 거같음 fill = 바탕색, eps= 모르겠음, ax= 낱개의 네모그래프, alpha=투명도
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)

    #산점도 그리기 ax=는 1, 3, 9 네모를 말함
    mglearn.discrete_scatter(X[:,0],X[:,1],y, ax=ax)
    
    #위쪽 제목 1,3,9 이웃출력
    ax.set_title("{}이웃".format(n_neighbors))
    #x축 설명 특성 0
    ax.set_xlabel("특성 0")
    #y축 설명 특성 1
    ax.set_ylabel("특성 1")
#첫번째 axes의 3번째 위치에 설명 출력
axes[0].legend(loc=3)
```
