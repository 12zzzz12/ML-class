```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import mglearn

#한글을 출력하기 위한 import
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname ="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)
import matplotlib as mpl # -깨짐 현상을 해결하기 위햏 matplotlib을 import
mpl.rcParams['axes.unicode_minus'] = False # - 깨짐 현상을 해결하기 위한 구문
```

```python
# iris 객체에 iris 데이터를 저장한다.
iris = datasets.load_iris()

# iris의 키 이름 들을 출력한다.
print(list(iris.keys()))

# iris의 타겟 이름들을 출력한다.
print("타깃의 이름 :{}".format(iris['target_names']))

# 특성의 이름을 출력한다. 
print("특성의 이름: {}".format(iris['feature_names']))

# iris 데이터의 인덱스3의 값만 X에 저장한다. 인덱스 3 = 'petal width'
X = iris["data"][:,3:]
print(X.shape)

# iris 타겟 인덱스=2의 타입을 y에 저장한다. 타겟 인덱스 2 = 'virginica'
# y= virginica이면 1 아니면 0
y = (iris["target"]==2).astype(np.int)
print(y.shape)
```

```python
# log_reg 객체에 로지스트 모델을 적용한다. 
log_reg = LogisticRegression()

# X,y를 학습시킨다. X=petal width 데이터, y= virginica이면 1 아니면 0
log_reg.fit(X,y)

#--------------------------
# linspace(start, stop, num) -> start ~ stop 까지 동일한 간격의 num구간을 생성
# num은 기본값으로 50을 준다.
# 즉 linspace(0,3,1000)은 0~3을 동일한 간격의 1000개로 나누는것

# reshape(a,b)는 (n,m) 행렬을 (a,b)로 변환 해줌 
# reshape(-1,1)은 그냥 (1,1) 행렬로 바꿔 주는듯?

# X_new = 0~3을 1000구역으로 나누어 일렬로 줄 세운 행렬임 [0,1,2, ..., 999]

X_new = np.linspace(0,3,1000).reshape(-1,1)

#--------------------------
# predict_proba = 각 클래스에 대한 확률값을 구해줌 (튜플, 2)
# 1번 원소는 첫번째 클래스일 확률, 2번 원소는 두번째 클래스일 확률 두개의 확률을 더하면 1
# ex) [0.4, 0.6] , [0.25, 0.75] 이런식으로 분류해주는 거 

# predict_proba(X_new) -> X_new값이 virginica인지 아닌지 확률값을 구해주는것
y_proba = log_reg.predict_proba(X_new)


#X = petal witdh값 , y = X 값에 따라 viriginca인지 아닌지 맞으면 1 아니면 0
plt.plot(X,y,"b.")


# y_proba에서 나눈 두번쨰 클래스 Virginca 일 확률값을 점으로 그림
plt.plot(X_new, y_proba[:,1],"g-",label="iris-Virginica")

# y_proba에서 나눈 첫번째 클래스 Not-Virginca일 확률값을 점으로 그림
plt.plot(X_new, y_proba[:,0],"b-",label="Not iris-Virginca")

# x,y 라벨의 설명을 출력한다.
plt.xlabel("Petal width",fontsize=14)
plt.ylabel("Probability",fontsize=14)

# 설명칸 위치 설정
plt.legend(loc=(0,1.05),fontsize=14)

```

```python
from sklearn.svm import LinearSVC
from sklearn.datasets import make_blobs

#X, y에 make_forge데이터를 저장한다.
X, y = mglearn.datasets.make_forge()

#전체 크기는 (10,3) 서브 그래프는 2개로 설정한다. 
fig, axes = plt.subplots(1,2,figsize=(10,3))

#2개의 서브 그래프를 for문을 이용하여 작성한다.
#LinearSVC , LogisticRegression 으로 작성한다.
for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    
    # clf객체를 생성하여 모델을 각각 LinearSVC , LogisticRegression로 하여 학습
    clf = model.fit(X,y)
    
    # 학습시킨 clf 객체의 X값을 점으로 이어 그린다. -> X값으로 선을 그리는거임
    # X, fill= 바탕색 여부, eps = 확대인듯? , ax= 그래프 구분, alpha = 그래프 선명도
    mglearn.plots.plot_2d_separator(clf,X,fill=False, eps=0.5, ax=ax, alpha=0.7)
    
    # 산점도를 그린다. 점들을 찍는거
    # X, y를이용해 점을 그린다.
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    
    # 각 그래프의 제목을 클래스를이용하여 작성한다.
    ax.set_title("{}".format(clf.__class__.__name__))
    
    # x,y 축의 라벨을 출력한다.
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
# 설명문 위치 설정
axes[0].legend()
```

```python
from sklearn.datasets import load_breast_cancer

#가져온 암 데이터를 cancer에 저장
cancer = load_breast_cancer()

#cancer데이터, 타겟을 훈련, 테스트 셋으로 나눈다.
X_train, X_test, y_train, y_test = train_test_split(
                        cancer.data, cancer.target, stratify=cancer.target,
                        random_state = 42)

#logreg 객체에 로지스트 모델을 만들어 훈련셋을 학습시킨다.
logreg = LogisticRegression(max_iter=5000).fit(X_train,y_train)

#훈련세트, 테스트 세트의 점수를 출력한다.
print("훈련 세트 점수: {:.3f}".format(logreg.score(X_train,y_train)))
print("테스트 세트 점수: {:.3f}".format(logreg.score(X_test,y_test)))
```

```python
# !!!C= 규제!!!
# C = 인벌스 만약 C가 100이라면 1/100을 곱한것과 같음 
# C=100 -> 즉 1/100 패널티가 작아진다.  -> 규제 완화 
# 개개의 포인터에 맞춤 -> 일반화가 아님 

#C=100인 로지스트 모델을 만들어 학습시킨다.
logreg100 = LogisticRegression(C=100,max_iter=5000).fit(X_train,y_train)

#C=100일 떄 훈련, 테스트 세트의 점수를 출력한다.
print("훈련 세트 점수: {:.3f}".format(logreg100.score(X_train,y_train)))
print("테스트 세트 점수: {:.3f}".format(logreg100.score(X_test,y_test)))

```

```python
#C=0.01인 로지스트 모델을 만들어 학습시킨다.
logreg001 = LogisticRegression(C=0.01,max_iter=5000).fit(X_train,y_train)

#C=0.01 떄 훈련, 테스트 세트의 점수를 출력한다.
print("훈련 세트 점수: {:.3f}".format(logreg001.score(X_train,y_train)))
print("테스트 세트 점수: {:.3f}".format(logreg001.score(X_test,y_test)))


# C= 0.01 이므로 100을 곱한것과 같음 
# 즉, 패널티가 커진다. -> 규제 강화
# 계수를 0에 가깝게
# 다수의 포인터에 맞추려고 함 -> 일반화 (훈련, 테스트 점수가 비슷해짐)
```

```python
# C 설명   [1/c] * 패널티 

#c가 작을 때 
# 복잡도 감소 , 성능 악화, 규제 강화, 계수를 0에 가깝게, 다수의 포인터

# c가 클 때
# 복잡도 증가, 성능 향상, 규제 완화, 훈련세트에 맞추려함 (일반화), 개개의 포인터

```

```python
#L1 규제

#C 값과 marker 모양을 변경해가며 for문을 진행한다.
for C, marker in zip([0.001, 1, 100], ['o','^','v']):
    
    #lr_l1 객체를 생성하여 패널티를 'l1'으로 하여 모델을 생성하고 학습을 진행한다.
    lr_l1 = LogisticRegression(solver='liblinear',C=C,penalty='l1').fit(X_train, y_train)
        
    #각 C의 값에 따른 로지스틱 모델의 훈련 정확도를 출력한다.
    print("C={:.3f}인 l1 로지스틱 회귀의 훈련 정확도: {:.2f}".format(
                      C, lr_l1.score(X_train, y_train)))
    
    #각 C의 값에 따른 로지스틱 모델의 테스트 정확도를 출력한다.
    print("C={:.3f}인 l1 로지스틱 회귀의 테스트 정확도: {:.2f}".format(
                     C, lr_l1.score(X_test, y_test)))
    
    #객체의 기울기에 해당되는 점을 찍는다.
    plt.plot(lr_l1.coef_.T, marker, label="C={:.3f}".format(C))

# xticks = x축의 눈금을 표시하는 함수
# range(cancer.data.shape[1]) = cancer 데이터의 피처 갯수만큼 눈금을 생성한다.
# cancer.feature_names = 생성된 눈금에 피쳐들의 이름을 집어 넣는다.
# rotation = 글씨 방향을 90도 돌린다.
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)

# x 범위 자동으로 맞춤
xlims = plt.xlim()

#수평선 그리기
# (y,x왼쪽끝, x 오른쪽끝) -> y=0에 xlims[0] ~ xlims[1]의 수평선을 그린다.
plt.hlines(0, xlims[0], xlims[1])

# x축 범위 자동맞춤하는듯?
plt.xlim(xlims)

# x, y의 라벨 출력
plt.xlabel("특성")
plt.ylabel("계수 크기")

# y축 범위 설정
plt.ylim(-5, 5)

# 설명칸 위치 설정
plt.legend(loc=3)
```

```python
print("사용한 특성의 개수:{}".format(np.sum(lr_l1.coef_ != 0)))
print("사용한 특성:{}\n".format(lr_l1.coef_))
```
