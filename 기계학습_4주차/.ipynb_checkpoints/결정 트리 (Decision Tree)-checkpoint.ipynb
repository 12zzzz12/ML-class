```python
#multi class -> a 아니면 나머지 , b 아니면 나머지  , (이진)
#(NB, KNN, DT, Logistic)

#mulitnomial -> a,b,c, 중 한가지를 선택'
from sklearn.model_selection import train_test_split
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import mglearn

#한글을 출력하기 위한 import
#한글 font를 설정해준다.

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname ="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

import matplotlib as mpl # -깨짐 현상을 해결하기 위햏 matplotlib을 import
mpl.rcParams['axes.unicode_minus'] = False # - 깨짐 현상을 해결하기 위한 구문

```

```python
#결정 트리 
# 분류 / 회귀에 사용

# 결정에 이르기 위해 가능한 적은 개수의 y/n 질문 사용 
# -> 트리의 레벨이 적을수록 좋음
# ex) 날개가 있나요? , 지느러미가 있나요?

# 과대적합 방지 방법
# 사전 가지치기 : 몇단계에서 중단 시켜버림
# 트리의 최대 깊이 제한, 리프의 최대 개수 제한

from sklearn.datasets import make_moons


#make_moons 
plt.title("make_moons : 초승달 모양 클러스터를 가진 가상 데이터")



#샘플 100개 , #noise = 잡음의 크기 0이면 반원을 이룸 
X, y = make_moons(n_samples = 100, noise = 0.25, random_state =3 )
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=100,
            edgecolor="k", linewidth=2)
```

```python
#2D Binary Classification Datasets

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_moons

#make_moons에서 가져온 데이터 X,y를 훈련, 테스트 셋으로 나눈다.
X_train , X_test, y_train, y_test = train_test_split(X, y,random_state =0, )

#tree 객체에 결정 트리 모델을 저장한다. #기본세팅 max_depth 없음
tree = DecisionTreeClassifier(random_state=0)


#tree 객체에 훈련 데이터 셋을 학습시킨다.
tree.fit(X_train, y_train)

#훈련, 테스트 셋의 정확도를 출력한다.
print("max_depth = default")
print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
#tree 객체에 결정 트리 모델을 저장한다. #최대깊이 제한을 4로 함
tree = DecisionTreeClassifier(max_depth=4,random_state=0)

#tree 객체에 훈련 데이터 셋을 학습시킨다.
tree.fit(X_train, y_train)

#훈련, 테스트 셋의 정확도를 출력한다.
print("max_depth = 4")
print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
#cancer 데이터로 결정트리 해봄
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
#cancer 데이터 저장
cancer = load_breast_cancer()

#cancer 데이터를 훈련 , 테스트 셋으로 나눈다.
X_train , X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                     stratify = cancer.target,
                                                     random_state =0 )
# X_train 형태를 출력
print("X_train.shape = {}".format(X_train.shape))

# 최대 깊이를 1, 4, 8로 해서 각 훈련, 테스트 정확도를 출력한다.
for max_depth in [1,4,8]:
    
    #tree 객체에 결정트리 모델을 적용하는데 max_depth를 각각 다르게 설정
    tree = DecisionTreeClassifier(max_depth = max_depth, random_state = 0)
    #tree 객체에 훈련셋 넣어서 학습
    tree.fit(X_train, y_train)
    
    #max_depth, 훈련, 테스트 정확도를 출력한다.
    print("max_depth = {}".format(max_depth))
    print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train,y_train)))
    print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
    print()
#최대 깊이가 커질수록 훈련 셋 
#정확도가 올라가는 것을 확인할 수 있음
```

```python
from sklearn.tree import export_graphviz

#graphviz 트리 출력
export_graphviz(tree, out_file = 'tree.dot', class_names=["악성", "양성"],
               feature_names = cancer.feature_names, impurity = False, filled=True)

import graphviz

with open("tree.dot", encoding= 'UTF8') as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

```python
#위 트리의 특성 중요도를 출력한다.
print("특성 중요도:\n{}".format(tree.feature_importances_))
```

```python
#해당 model의 중요도를 출력하는 메소드
def plot_feature_importances_cancer(model):
    
    #cancer의 피처(30개)들을 n_feature에 저장한다.
    n_features = cancer.data.shape[1]
    
    # barh = 가로막대 그리는 함수
    # range()=  범위(개수 말하는듯?) , model.feature의 중요도, 간우데 맞춤
    plt.barh(range(n_features), model.feature_importances_, align='center')
    
    #yticks = y축 눈금값 설정 
    # arange() n_feature들의 개수, cancer의 피처이름들을 축에 적음
    plt.yticks(np.arange(n_features), cancer.feature_names)
    
    #x, y라벨의 이름 출력
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    
    #y축의 범위 1만큼의 길이에 n_features로 나눔
    plt.ylim(-1, n_features)

#tree의 중요도를 출력
plot_feature_importances_cancer(tree)

# 0 ~ 1 사이의 숫자를 출력
# 0 = 전혀 사용 x , 1 = 타깃 클래스 완벽하게 예측
# 전체 합 = 1

#중요도가 낮다고 해서 유용하지 않은 것은 아님
# 중요도 = 양수

# 어떤 클래스를 지지하는지는 모름
```

```python
#결정 트리 회귀
import os
#ram_price 변수에 csv데이터 정보를 저장한다.
ram_prices = pd.read_csv(os.path.join(mglearn.datasets.DATA_PATH,'ram_price.csv'))

#ramprice의 형태를 출력한다.
print("ram_prices.shape: {}".format(ram_prices.shape))

#ramprice의 key를 출력한다.
print("ram_prices.keys: {}".format(ram_prices.keys()))

#ram_price의 ~5번째까지 데이터를 출력한다.
print("{}".format(ram_prices[:5]))


#print("ram_prices.date: {}".format(ram_prices.date))
#print("ram_prices.price: {}".format(ram_prices.price))

# y축의 폰트를 Arial로 설정한다.
plt.yticks(fontname = 'Arial')

#semilogy(x,y) x 축은 ram_prices.date 날짜를 출력한다.
#              y 축은 ram_prices.price를 10의 거듭제곱으로 표현해서 출력한다.
plt.semilogy(ram_prices.date, ram_prices.price)

#x, y 라벨의 제목을 출력한다.
plt.xlabel("년")
plt.ylabel("가격 ($/Mbyte)")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

#훈련 데이터 = ram_price의 날짜가 2000년도 이전인것
data_train = ram_prices[ram_prices.date<2000]

#테스트 데이터 = ram_price의 날짜가 2000년도 이후인것
data_test = ram_prices[ram_prices.date>=2000]

#훈련, 테스트 데이터의 형태들을 출력
#훈련 -> 202개 ,  테스트 -> 131개
print("data_train.shape: {}".format(data_train.shape), 
      'data_test.shape:{}'.format(data_test.shape))

# newaxis -> 없는 배열을 하나 만들어줌 ex) (4,) -> (4,1)
#print("data_train.shape: {}".format(data_train.date.shape)) 
#위에 해보면 (202,)로 출력되는 걸 확인가능
X_train = data_train.date[:, np.newaxis]
#즉, X_traindp (202,1)로 만들어 다시 저장해주는것임
#print("data_train.shape: {}".format(data_train.date.shape)) 

# np.log(x) -> 밑이 e인 로그를 x에 씌우고 y_train에 저장
y_train = np.log(data_train.price)

#tree 객체에 결정트리 모델을 넣고 훈련 데이터를 학습시킨다.
#한줄이지만 안에서는 많은 계산이 돌아가는 중임
tree = DecisionTreeRegressor().fit(X_train, y_train)

#linear_reg 객체에 선형회귀 모델을 넣고 훈련 데이터를 학습시킨다.
linear_reg = LinearRegression().fit(X_train, y_train)

#date를 (202,) 에서 (202,1) 로 만들어 X_all에 저장한다.
X_all = ram_prices.date[:, np.newaxis]

#X_all 즉, 결정트리로 구한 ram_prices.date의 예측값을 pred_tree에 저장한다.
pred_tree = tree.predict(X_all)

#X_all 즉, 선형회귀로 구한 ram_prices.date의 예측값을 pred_lr에 저장한다.
pred_lr = linear_reg.predict(X_all)

#np.exp(x) -> 밑이 e인 지수함수로 바꿔줌
#ex) np.exp(2) -> e^2
price_tree = np.exp(pred_tree)
price_lr = np.exp(pred_lr)

```

```python
#y 축의 폰트를 Arial로 설정한다.
plt.yticks(fontname = "Arial")

#x축은 train.date, y축은 train.price를 10의 거듭제곱으로 표현
plt.semilogy(data_train.date, data_train.price, label = "훈련 데이터")

#x축은 test.date, y축은 test.price를 10의 거듭제곱으로 표현
plt.semilogy(data_test.date, data_test.price, label = "테스트 데이터")


#x축은 price.date, y축은 price_tree = exp(pred_tree)를 10의 거듭제곱으로 표현
plt.semilogy(ram_prices.date, price_tree, label = "트리 예측")

#x축은 price.date, y축은 pirce_lr = exp(pred_lr)를 10의 거듭제곱으로 표현
plt.semilogy(ram_prices.date, price_lr, label = "선형회귀 예측")

#설명값 출력
plt.legend()


#결정 트리 회귀 분석

#트리 예측 보면 외삽(훈련 데이터 범위 밖을 예측하는 것)
#시계열 데이터에 잘 맞지 않음
#등락의 예측에는 적용 가능 

```

```python
#결정트리 장단점

#모델 복잡도 제어 매개변수가 있음
#max_depth : 트리 최대 깊이
#max_leaf_nodes : 리프 노드의 최대 개수
#min_samples_leaf : 리프 노드가 되기 위한 최소 샘플 개수
#min_samples_split : 노드가 분할하기 위한 최소 샘플 개수

#트리가 작으면 시각화가 좋아 설명이 쉬움

#특성이 각각 처리됨
# -> 데이터 스케일에 구애 받지 않음
# 정규화나 표준화 같은 전처리 과정이 필요없음
# 이진 특성이나 연속적인 특성이 혼합되어 있어도 가능


#단점: 과대 적합 가능
```

```python
#붓꽃 결정트리로 분류하기!!!!

from sklearn.datasets import load_iris
from graphviz import Source

#iris에 가져온 iris데이터를 저장한다.
iris = load_iris()

X = iris.data[:, 2:] #꽃잎 길이와 너비
y = iris.target #꽃잎의 타겟 

#tree_clf 객체를 생성한다. 최대 깊이 = 2로 설정한다.
tree_clf = DecisionTreeClassifier(max_depth = 2, random_state = 42)

#꽃잎 데이터(x) 와 target(y) 값으로 tree_clf를 학습시킨다.
tree_clf.fit(X,y)

# graphviz를 설정한다.
# 모델 = tree_clf
# out_file = 결과값을 해당 파일에 저장한다. -> "iris_tree.dot"에 결과 저장
# class_names = 최종결정된 클래스의 이름 
# round = 테두리 부분을 둥글게
# filed = 바탕색을 칠함
export_graphviz(tree_clf, out_file = "iris_tree.dot",
               feature_names = iris.feature_names[2:],
               class_names = iris.target_names,
               rounded = True,
               filled = True)

Source.from_file("iris_tree.dot")




```

```python
from sklearn.tree import plot_tree

#도표 크기를 (10,8)로 설정한다.
plt.figure(figsize=(10,8))

#모델을 gini가 아닌 entropy로 결정하도록 한다.
tree_clf = DecisionTreeClassifier(criterion='entropy', random_state = 42)

#그다음 모델에 X,y를 학습시킨다.
tree_clf.fit(X,y)

#plot_tree
#모델 = tree_clf
# filled = 바탕색 
# rounded = 테두리 둥글게?
# max_depth = 최대 깊이
#feature_naems = 피처들의 이름 (iris.feature_names[2:]) iris피처 앞 2개
#여기서는 petal length와 peatl width를 사용
#class_names = iris.target의 이름을 사용
plot_tree(tree_clf, filled = True, rounded = True,
         max_depth = 2,
         feature_names = iris.feature_names[2:],
         class_names = iris.target_names)

#도표 출력
plt.show()
```
