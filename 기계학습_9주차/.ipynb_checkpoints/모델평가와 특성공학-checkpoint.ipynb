```python
import mglearn

#k 겹 교차 검증
#k=5 or k=10
# 훈련 데이터를 k개의 부분 집합으로 분할
# 첫번째 폴드를 테스트 세트로하고 나머지 폴드로 모델을 훈련
# 테스트 폴드를 바꾸어 가면 모든 폴드가 사용될때까지 반복
mglearn.plots.plot_cross_validation()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

#sk- learn 교차 검증
iris = load_iris()
logreg = LogisticRegression()

#교차검증 점수를 출력한다.
scores = cross_val_score(logreg, iris.data, iris.target)
print("교차 검증 점수: {}".format(scores))
```

#교차 검증 장단점

train_test_split 
- 무작위로 데이터를 나누기 때문에 우연에 따라 평가 결과가 달라질 수 있음

교차검증

- 모든 폴드가 테스트 대상이 되기 때문에 공평함
- 10겹 교차 검증은 90퍼를 훈련에 사용하기 때문에 비교적 정확한 평가
- 모델이 훈련 데이터에 대한 민감도 파악
- 최악의 경우와 최선의 경우를 예측 

단점
연산 비용의 증가

```python
mglearn.plots.plot_stratified_cross_validation()

#순서대로 나열된 레이블에 적용한 기본 교차 검증
#클래스 0 테스트, 나머지 훈련으로 하면 가끔 성능이 안좋아지는 경우가 있음

#계층별 교차 검증
#각각의 클래스에서 조금씩 테스트 셋을 가져와서 훈련하면 
#데이터를 골고루 사용하므로 성능이 확실하다.
```

```python
from sklearn.model_selection import KFold

#kfold객체에 KFold모델을 적용한다.
kfold = KFold(n_splits=5)

#kfold 객체를 cross_val_score를 통해 점수를 확인한다.
print("교차 검증 점수:\n{}".format(cross_val_score(logreg, iris.data, iris.target,
                                            cv=kfold)))
```

```python
mglearn.plots.plot_shuffle_split()

#10개의 데이터 포인트에 train_size=5, test_size=2와 n_splits=4를 적용한
#shufflesplit을 그린다.
```

```python
from sklearn.model_selection import ShuffleSplit
shuffle_split = ShuffleSplit(test_size=.5, train_size=.5, n_splits=10)
scores = cross_val_score(logreg, iris.data, iris.target, cv=shuffle_split)
print("교차 검증 점수:\n{}".format(scores))
```

```python
#그룹별 교차 검증

데이터 내에 연관된 그룹이 있을 경우
타깃에 따라 폴드를 나누지 않고 입력 특성에 따라 폴드를 나누어야 할 경우

데이터 : 100장의 사진 데이터
목적 : 사람의 표정을 분류
    
데이터 특성
- 한사람의 다른 표정이 여러장 포함되어 있음
- 한사람의 데이터 훈련, 테스트 모두 나타날 경우 분류기 성능 정확히 측정 불가
```
