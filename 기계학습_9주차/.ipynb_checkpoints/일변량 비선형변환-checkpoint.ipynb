```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import mglearn
#한글을 출력하기 위한 import
#한글 font를 설정해준다.
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname ="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

import matplotlib as mpl # -깨짐 현상을 해결하기 위햏 matplotlib을 import
mpl.rcParams['axes.unicode_minus'] = False # - 깨짐 현상을 해결하기 위한 구문
```

```python
#랜덤 숫자를 rnd에 적용한다.
rnd = np.random.RandomState(0)
X_org = rnd.normal(size=(1000,3))
w = rnd.normal(size=3)

#X값에 포이즌을 적용해서 숫자들을 설정한다.
X = rnd.poisson(10 * np.exp(X_org))
#y값에 점을 찍어 w를 설정한다.
y = np.dot(X_org, w)
print(X[:10, 0])
print("특성 추현 횟수:\n{}".format(np.bincount(X[:,0])))
```

```python
#위에서 구한 특성 출현횟수 와 값들을 그래프로 그림
bins = np.bincount(X[:, 0])
plt.bar(range(len(bins)), bins, color='grey')
plt.ylabel('출현 횟수')
plt.xlabel('값')
```

```python
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

#cancer 데이터 노이즈를 추가한다.
noise = rnd.normal(size=(len(cancer.data), 50))

#hstack을 통해 X_w_noise 데이터를 생성한다.
X_w_noise = np.hstack([cancer.data, noise])

#X_w_noise, caner.target을 이용해서 훈련, 테스트 셋을 만든다.
X_train, X_test, y_train, y_test = train_test_split(X_w_noise, cancer.target
                                                   , random_state = 0 , test_size=.5)

#selectpercentile에서 select 객체를 생성한다.
select = SelectPercentile(score_func=f_classif, percentile=50)

#생성한 객체에서 훈련셋을 학습시킨다.
select.fit(X_train, y_train)
X_train_selected = select.transform(X_train)

#mask와 해당 amskㄱ의 그래프를 그린다.
mask = select.get_support()
print(mask)
plt.matshow(mask.reshape(1,-1), cmap='gray_r')
plt.xlabel("특성 번호")
```

```python
from sklearn.linear_model import LogisticRegression

#테스트 셋을 select객체를 적용하여 테스트 셀렉 데이터를 생성한다.
X_test_selected = select.transform(X_test)

#로지스틱회귀를 생성하고 훈련셋을 학습시킨다.
lr = LogisticRegression()
lr.fit(X_train, y_train)

#훈련시킨 로지스틱의 점수를 출력한다.
print("전체 특성을 사용한 점수: {:.3f}".format(lr.score(X_test, y_test)))

#셀렉된 훈련데이터셋을 학습시킨 뒤에 점수를 출력한다.
lr.fit(X_train_selected, y_train)
print("선택된 일부 특성을 사용한 점수: {:.3f}".format(lr.score(X_test_selected, y_test)))
```

```python
#반복적 특징 선택
#특성을 하나씩 추가하면서 모델을 만든다.
#모든 특성에서 하나씩 제거하면서 모델을 만든다.
#특징의 수가 다른 여러 개의 모델이 생성되어 계산 비용이 많이 든다.

#RFE, 랜덤 포레스트 라이브러리 IMPORT
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

#RFE모델을 집어넣을 select 객체를 생성한다.
select = RFE(RandomForestClassifier(n_estimators=100, random_state=42),
            n_features_to_select = 40)

#select 객체에 훈련셋을 학습시킨다.
select.fit(X_train, y_train)

#select 객체를 이용해 mask 생성
mask = select.get_support()

#mask 그래프로 출력한다.
plt.matshow(mask.reshape(1,-1), cmap='gray_r')
plt.xlabel('특성 번호')
```

```python
#전문가 지식활용

#자전거 데이터를 citibike에 저장한다.
citibike = mglearn.datasets.load_citibike()

#그래프를 그릴 구간 설정
plt.figure(figsize=(10,3))

# x축의 이름을 설정한다.
xticks = pd.date_range(start=citibike.index.min(), end = citibike.index.max(),
                      freq = 'D')
week = ["일",'월','화','수','목','금','토']
xticks_name = [week[int(w)] + d for w, d in zip(xticks.strftime('%w'),
                                               xticks.strftime('%m - %d'))]

#자전거 그래프를 출력
plt.xticks(xticks, xticks_name, rotation=90, ha = 'left')
plt.plot(citibike, linewidth=1)
plt.xlabel('날짜')
plt.ylabel('대여횟수')
```

```python
#시간과 요일 특성을 사용해 만든 랜덤포레스트 예측
from sklearn.ensemble import RandomForestRegressor

n_train = 184

#X, y에 시티바이크 값들을 저장한다.
y = citibike.values
X = citibike.index.astype('int64').values.reshape(-1,1)

#특성에 따른 그래프를 그려주는 함수
def eval_on_features(features, target, regressor):
    X_train, X_test = features[:n_train], features[n_train:]
    
    y_train, y_test = target[:n_train], target[n_train:]
    
    #위에서 나눈 훈련 셋을 학슶시킨다.
    regressor.fit(X_train, y_train)
    
    #regressor 객체의 모델 예측값을 출력한다.
    print("테스트 세트 R2:{:.2f}".format(regressor.score(X_test,y_test)))
    
    #y_pred에 regressor 예측값들을 저장한다.
    y_pred = regressor.predict(X_test)
    y_pred_train = regressor.predict(X_train)
    
    #그래프를 그리기 위해 크기 설정
    plt.figure(figsize=(10,3))
    plt.xticks(range(0,len(X),8), xticks_name,rotation=90,ha='left')
    
    #훈련, 테스트, 훈련예측, 테스트예측 4가지 선을 그리기 위해 설정한다.
    plt.plot(range(n_train), y_train, label="훈련")
    plt.plot(range(n_train,len(y_test)+n_train), y_test, '-', label='테스트')
    plt.plot(range(n_train),y_pred_train, '--', label="훈련 예측")
    plt.plot(range(n_train, len(y_test)+n_train),y_pred,'--',
            label="테스트 예측")
    
    #설명칸과 x,y축 라벨 설정
    plt.legend(loc=(1.01,0))
    plt.xlabel('날짜')
    plt.ylabel('대여횟수')
    
#랜덤포레스트로 regressor 객체에 저장한다.
regressor = RandomForestRegressor(n_estimators=100, random_state=0)

#날짜와 요일을 이용하여 feature를 설정한다.
X_hour_week = np.hstack([citibike.index.dayofweek.values.reshape(-1,1),
                        citibike.index.hour.values.reshape(-1,1)])
eval_on_features(X_hour_week,y,regressor)
```
